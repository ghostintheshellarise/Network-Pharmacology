```python
drugs=df_drug['Drug_description']
```

```python
nlp = spacy.load("en_core_sci_lg")
```

```python
# Test case
token = nlp('Apitolisib')
print(token.text, token.has_vector, token.vector_norm)
```

```python
token.vector
```

```python
for entry in drugs:
    token = nlp(entry)
    print(token.text, token.has_vector, token.vector_norm)
```

```python
start_nodes = df_network['start_node']
```

```python
for entry in list(set(start_nodes)):
    token = nlp(entry)
    print(token.text, token.has_vector, token.vector_norm)
```

```python
from scipy.spatial.distance import cosine
```

```python
import time
start_time = time.time()
drug_node_pair_similarity = list()
for drug in drugs:
    drug_token = nlp(drug)
    if drug_token.vector_norm!=0:
        for node in start_nodes:
            node_token = nlp(node)
            if node_token.vector_norm!=0:
                score = cosine(drug_token.vector, node_token.vector)
                drug_node_pair_similarity.append([drug,node,score])
print ("Total CPU runtime: %.4f" %(time.time()-start_time))
```

```python
for drug_node_pair in drug_node_pair_similarity:
    print (drug_node_pair)
```

```python
df = pd.DataFrame(drug_node_pair_similarity, columns=['drug','node','cosine'])
```

```python
df.head()
```

```python
print("Total number of pairs are %i" %(len(df)))
```

```python
print ("Min score: %.4f" %(min(df['cosine'])))
print ("Max score: %.4f" %(max(df['cosine'])))
```

```python
from IPython.display import IFrame
IFrame(src='https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.spatial.distance.cosine.html', width=600, height=400)
```

```python
shortlisted_drugs = list()
for drug,node,cosine in drug_node_pair_similarity:
    if cosine<0.5:
        shortlisted_drugs.append([drug,node,cosine])
```

```python
print(shortlisted_drugs) # Seems there are duplicates.
```

```python
effective_drug_node_pairs = list()
for pairs in shortlisted_drugs:
    if pairs not in effective_drug_node_pairs:
        effective_drug_node_pairs.append(pairs)
    else:
        pass
```

```python
print(effective_drug_node_pairs)
print (len(effective_drug_node_pairs))
```

```python
# Extract text to identify the kind of relationship between drug-node pair
# test case with 'Interferon gamma-n1' and 'TNF'
import time
from Bio import Medline
from Bio import Entrez
Entrez.email="simonkamantse@outlook.com"
handle = Entrez.esearch(db="pubmed", term="Interferon gamma-n1 AND TNF")
record = Entrez.read(handle)
record["IdList"]

def remove_heading(txt):
    pos = txt.find(":")
    if pos==-1:
        return txt
    return txt[pos+2:]

globals
acc_list = record['IdList']
batch_size = 200
count = len(acc_list)
out_lst=list()
start_time = time.time()
for start in range(0, count, batch_size):
    end = min(count, start+batch_size)
    print("Going to download record %i to %i" % (start+1, end))
    id_list = acc_list[start:end]
    fetch_handle = Entrez.efetch(db="pubmed",id=acc_list,rettype="medline",retmode="text",retmax=batch_size)
    records=Medline.parse(fetch_handle)
for record in records:
    out_lst.append(remove_heading(record.get("AB","?")))
print ("Total CPU runtime : %.4f" %(time.time()-start_time))
```

```python
for abstract in out_lst:
    if "Interferon gamma-n1" in abstract or "TNF" in abstract:
        print(abstract)
        print()
```
